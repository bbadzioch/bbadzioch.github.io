## Groupby.apply

```python
def large_tips(df):
    df = df.sort_values(by="tip", ascending=False)
    return df.head(3)

big_tips = grouped.apply(large_tips)
big_tips
```

## Combining DataFrames

## Merging

```python
instructors = pd.read_csv(url + "instructors.csv")
instructors
```

```python
courses = pd.read_csv(url + "courses.csv")
courses
```

```python
schedule = pd.merge(courses, instructors, left_on="instructor", right_on="name", how="inner")
schedule
```

```python
schedule.drop("name", axis=1)
```

```python
pd.merge(courses, instructors, left_on="instructor", right_on="name", how="left")
```

```python
pd.merge(courses, instructors, left_on="instructor", right_on="name", how="right")
```

```python
pd.merge(courses, instructors, left_on="instructor", right_on="name", how="outer")
```

## Plotting with Seaborn

```python
import seaborn as sns

sns.relplot(data=tips,
            x="total_bill",
            y="tip"
);
```

```python
sns.set_theme(style="darkgrid",
              palette="deep",
              font_scale=1.3
             )

sns.relplot(data=tips,
            x="total_bill",
            y="tip"
);
```

```python
sns.relplot(data=tips,
            x="total_bill",
            y="tip",
            hue="day",
            height=6,
            aspect=2
);
```

```python
sns.relplot(data=tips,
            x="total_bill",
            y="tip",
            hue="day",
            size="size",
            sizes=(5, 400),
            height=6,
            aspect=2
);
```

```python
sns.relplot(data=tips,
            x="total_bill",
            y="tip",
            hue="day",
            size="size",
            col="time",
            sizes=(5, 400),
            height=10,
            aspect=1
);
```

```python
sns.relplot(data=tips,
            x="total_bill",
            y="tip",
            hue="day",
            size="size",
            col="time",
            row="sex",
            sizes=(5, 400),
            height=10,
            aspect=1
);
```

```python
sns.barplot(data=tips, 
            x="day", 
            y="total_bill",
            estimator=sum, 
            order=["Thur", "Fri", "Sat", "Sun"]
           );
```

## Plotting with Plotly Express

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook"
```

```python
fig = px.scatter(data_frame=tips, 
                 x="total_bill",
                 y="tip",
                 title="Restaurant data",
                 width=800,
                 height=400
                )
fig.show()
```

```python
fig = px.scatter(data_frame=tips, 
                 x="total_bill",
                 y="tip",
                 color="day",
                 title = "Restaurant data"
                )
fig.show()
```

```python
fig = px.scatter(data_frame=tips, 
                 x="total_bill",
                 y="tip",
                 color="day",
                 size="size",
                 title = "Restaurant data"
                )
fig.show()
```

```python
fig = px.scatter(data_frame=tips, 
                 x="total_bill",
                 y="tip",
                 color="day",
                 size="size",
                 facet_col="time",
                 title = "Restaurant data"
                )
fig.show()
```

```python
fig = px.scatter(data_frame=tips, 
                 x="total_bill",
                 y="tip",
                 color="day",
                 size="size",
                 facet_col="time",
                 facet_row="sex",
                 title = "Restaurant data",
                 width=800,
                 height=800
                )
fig.show()
```

```python
fig = px.bar(tips,
             x="day",
             y="tip",
             color="sex",
             barmode="group", # relative group overlay
             category_orders = {"day" : ["Thur", "Fri", "Sat", "Sun"]}, # order days on x-axis
             title="Bar plot"
            )
fig.show()
```

```python
fig = px.histogram(tips,
                   x="total_bill",
                   labels = {"total_bill" : "total bill"},
                   nbins = 6,
                   title = "Histogram"
                  )

fig.update_layout({"bargap": 0.02})  # add space between bars

fig.show()
```
