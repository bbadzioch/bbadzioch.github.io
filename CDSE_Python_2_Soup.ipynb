<h1 class="bbh">
3. Beautiful Soup
</h1>


<hr/>

<a href="https://www.crummy.com/software/BeautifulSoup/">Beautiful Soup</a> is a Python library for parsing HTML documents. 

</div>

```python
from bs4 import BeautifulSoup
```

```python
grads
```

<div class="bb alert alert-info">
Parse the page:
</div>

```python
soup = BeautifulSoup(grads)
```

<div class="bb alert alert-info">
Print the page code neatly:
</div>

```python
print(soup.prettify())
```

<div class="bb alert alert-info">
Find all HTML <code>&lt;p&gt;...&lt;/p&gt;</code> tags:
</div>

```python
ptags = soup.find_all('p')
ptags
```

<div class="bb alert alert-info">

<p>The <code>get_text()</code> method returns text enclosed in a tag.</p>
</div>

```python
ptags[1]
```

```python
ptags[1].get_text()
```

<div class="bb alert alert-info">
Get text from all paragraphs containing the word "Office":
</div>

```python
grad_list = [t.get_text() for t in ptags if "Office" in t.get_text()]
grad_list
```

<h1 class="bbh">
4. String Processing
</h1>
    
<hr/>

Here we get names and emails from the strings.

</div>

```python
dustin = grad_list[10]
dustin
```

```python
n = dustin.find("Office")
n
```

```python
dustin[0:n].strip()
```

```python
def get_name(s):
    stop = s.find("Office:")
    return s[:stop].strip()
```

```python
get_name(dustin)
```

```python
def get_email(s):
    start = s.find("Email:") + 6
    return s[start:].strip()

get_email(dustin)
```

```python
for s in grad_list:
    name = get_name(s)
    email = get_email(s)
    print(f"Name:   {name}")
    print(f"Email:  {email}")
    print()
```
